```python
%matplotlib inline
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
 
```

```python
fig = plt.figure(figsize = (10, 10))
x = np.linspace(-5, 5, 100)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1.5, 20) 
plt.grid(True)
plt.show()
def equations(p):
    x, y = p
    return ( y - x**2 + 1, np.exp(x) + x - x*y - 1)
x1, y1, =  fsolve(equations, (-2, -1))
x2, y2, =  fsolve(equations, (2, 3))
x3, y3, =  fsolve(equations, (4, 5))
print (round(x1, 2), round(y1, 2))
print (round(x2, 2), round(y2, 2))
print (round(x3, 2), round(y3, 2))
#(0; 1) вроде тоже корни))
```
