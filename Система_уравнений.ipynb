```python
%matplotlib inline
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
 
```

```python
x = np.linspace(-2, 7, 100)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1.5, 25) 
plt.grid(True)
plt.show()
def equations(p):
    x, y = p
    return ( y - x**2 + 1, np.exp(x) + x - x*y - 1)
x1, y1, =  fsolve(equations, (-2, 6))
print (round(x1, 2), round(y1, 2))
```
