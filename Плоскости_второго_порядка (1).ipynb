```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize = (5, 5))
ax = Axes3D(fig)
ax.set_title('Две_параллельные_плоскости_второго_порядка')
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
x, y = np.meshgrid(x, y)
z1 = (6-3*x**2 - 4*y)/5
z2 = (-20-6*x**2 - 8*y)/10
surf = ax.plot_surface(x, y, z1, rstride=1, cstride=1, cmap=cm.plasma)
surf = ax.plot_surface(x, y, z2, rstride=1, cstride=1, cmap=cm.plasma)
show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize = (5, 5))
ax = Axes3D(fig)
ax.set_title('Плоскость_второго_порядка')
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
x, y = np.meshgrid(x, y)
z = np.sqrt(3*x**2 + 4*y**2)
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.plasma)
show()
```
