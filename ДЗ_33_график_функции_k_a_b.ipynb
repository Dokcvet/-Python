```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
fig = plt.figure(figsize = (10, 10))
k = [1, 10, 20]
a = [2, 4, 6]
b = [3, 5, 7]
x = np.linspace(-10, 10, 100)
y = k[0]*np.cos(x-a[0])+b[0]
y1 = k[1]*np.cos(x-a[1])+b[1]
y2 = k[2]*np.cos(x-a[2])+b[2]
plt.plot(x, y)
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```
