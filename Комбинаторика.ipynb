```python
import numpy as np
import itertools
```

```python
for p in itertools.product("0011",repeat=6):
    print(''.join(p)) 
```

```python
for p in itertools.permutations("qwerty", 4):
    print(''.join(str(x) for x in p)) 
```

```python
for p in itertools.combinations("qwerty", 4):
    print(''.join(p)) 
```
